```python
import numpy as np
```

```python
lammin = 600
lammax = 900
R = 150
```

```python
loglammin, loglammax = np.log([lammin, lammax])
dloglam = 1/R
# int(np.ceil((loglam_max - loglam_min)/dloglam))+1?
nlam = int(np.ceil((loglammax - loglammin) / dloglam) + 1)

print(nlam)
```

```python
loglams = np.arange(nlam) / (nlam-1) * (loglammax - loglammin) + loglammin
loglams
```

```python
np.diff(loglams)
```

```python
lams = np.exp(loglams)
```

```python
lams[1:] / np.diff(lams)
```

```python
lams = [lammin]
for i in range(nlam-1):
    lam = lams[i] + lams[i]/R
    lams.append(lam)
lams = np.array(lams)
```

```python
lams
```

```python
lams[:-1]/np.diff(lams)
```

```python
np.diff(np.log(lams))
```
